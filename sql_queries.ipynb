```python
import pandas as pd
import sqlite3
from datetime import datetime as dt
import os
```

```python
caminho_db = 'c:/Users/rafae/PIPELINE/projeto_pipeline/pipeline/pipeline.db'

# Estabeleça a conexão
conn = sqlite3.connect(caminho_db)
print("Conexão com banco de dados estabelecida")
```

QUERY 1 : VISÃO GERAL DOS DADOS 

```python
query = """
SELECT COUNT(*) as total_registros 
FROM tabela_principal
"""

resultado = pd.read_sql_query(query, conn)
print("Total de registros:", resultado['total_registros'].values[0])
```

QUERY 2 : TOP 10 (USUÁRIOS COM MAIS POST)

```python
query = """
SELECT 
    usuario,
    COUNT(*) as total_posts
FROM 
    tabela_principal
GROUP BY 
    usuario
ORDER BY 
    total_posts DESC
LIMIT 10
"""

top_usuarios = pd.read_sql_query(query, conn)
print("\nTop 10 Usuários que mais postaram:")
print(top_usuarios)
```

QUERY 3 : ANÁLISE TEMPORAL

```python
query = """
SELECT 
  STRFTIME('%H', data_ingestao) as hora,
  COUNT(*) as total_postagens
FROM tabela_principal
GROUP BY hora
ORDER BY hora
"""

postagens_por_hora = pd.read_sql_query(query, conn)

print("\nPostagens por hora:")
print(postagens_por_hora)

conn.close()
```
